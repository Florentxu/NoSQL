```python
from pymongo import MongoClient
from pprint import pprint, PrettyPrinter
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.gym
gymnases = db.Gymnases
sportifs = db.Sportifs
print('connexion success')
```

## 1) Afficher toutes les collections de la base, Compter le nombre de documents ds chacune des collections, Afficher un document de chacune des collections ?

```python
#Afficher toutes les collections de la base
print("Toutesl es collesctions de la base gym")
print(db.collection_names())

#Compter le nombre de documents ds chacune des collections
nbg = gymnases.count()
nbs = sportifs.count()

print(nbg," documents dans la collections Gymnases")
print(nbs," documents dans la collections Sportifs")

#Afficher un document de chacune des collections
print("un document de chacune de la collections Gymnases")
pprint(gymnases.find_one())

print("un document de chacune de la collections Sportifs")
pprint(sportifs.find_one())
```

## 2) On cherche le (ou les) conseiller(s) s’appelant "KERVADEC" ?

```python
for conseiller in sportifs.find({"Nom":"KERVADEC"}):
    pprint(conseiller,sort_dicts=False)
```

## 3) On cherche les sportifs qui ont ou bien 32 ans ou 40 ans ?

```python
for sportif in sportifs.find({"$or":[{"Age":32}, {"Age": 40}]}):
    pprint(sportif,sort_dicts=False)
```

## 4) On cherche les sportifs jouant au Basket-ball ?

```python
for sportif in sportifs.find({"Sports.Jouer":"Basket ball"}):
    pprint(sportif,sort_dicts=False)
```

## 5) On cherche les sportifs ayant au moins 32 ans, ou de sexe féminin ?

```python
for sportif in sportifs.find({"$or": [{"Age":{"$gte":32}},{"Sexe":"F"}]}):
    pprint(sportif,sort_dicts=False)
```

## 6) Trouver le nombre de sportifs de sexe féminin dans la base ?

```python
nbf = sportifs.count({"Sexe":{"$in":["F","f"]}})
print(nbf, "femmes sportifs")
```

## 7)	Quels sont les sportifs (identifiant, nom et prénom) qui ont entre 20 et 30 ans ?

```python
for sportif in sportifs.find({"Age":{"$gte":20,"$lte":30}},{"_id":0,"IdSportif":1,"Nom":1,"Prenom":1,"IdSportifConseiller":1}):
    pprint(sportif,sort_dicts=False)
```

## 8)	Quels sont les sportifs (identifiant et nom) qui pratiquent du hand ball ?

```python
for sportif in sportifs.find({"Sports.Jouer":"Hand ball"},{"_id":0,"IdSportif":1,"Nom":1}):
    pprint(sportif,sort_dicts=False)
```

## 9)	Quels sont les gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2 ?

```python
for gymnase in gymnases.find({"$or":[{"Ville":"VILLETANEUSE"},{"Ville":"SARCELLES"}],"Surface":{"$gt":400}}):
    pprint(gymnase,sort_dicts=False)
```

## 10)	Dans quels gymnases peut-on jouer au hockey le mercredi après 15H ?

```python
for gymnase in gymnases.find({"Seances.Jour":{"$in":["Mercredi","mercredi"]},"Seances.Libelle":"Hockey","Seances.Horaire":{"$gte":15}}):
    pprint(gymnase,sort_dicts=False)
```

## 11)	Quels sportifs (identifiant et nom) ne pratiquent aucun sport ?

```python
for sportif in sportifs.find({"Sports.Jouer":{"$exists":False}},{"_id":0,"IdSportif":1,"Nom":1}):
    pprint(sportif,sort_dicts=False)
```

## 12)	Quels gymnases n’ont pas de séances le dimanche ?

```python
for gymnase in gymnases.find({"Seances.Jour":{"$nin":["Dimanche","dimanche"]}}):
    pprint(gymnase,sort_dicts=False)
```

## 13)	Quels gymnases ne proposent que des séances de basket ball ou de volley ball ?

```python
for gymnase in gymnases.find({"Seances.Libelle":{"$all":["Basket ball","Volley ball"]}}):
    pprint(gymnase,sort_dicts=False)
```

## 14)	Quels sont les entraîneurs qui sont aussi joueurs ?

```python
for sportif in sportifs.find({"Sports.Entrainer":{"$exists":True},"Sports.Jouer":{"$exists":True}}):
    pprint(sportif,sort_dicts=False)
```

## 15)	Quels sont les sportifs qui sont des conseillers ?

```python
arrayConseiller = []
conseillers = sportifs.find({"IdSportifConseiller": {"$exists": True}}, {"IdSportifConseiller": 1, "_id": 0}).sort("IdSportifConseiller", 1)
for conseiller in conseillers:
    if conseiller not in arrayConseiller:
        arrayConseiller.append(conseiller)
        request = sportifs.find_one({"IdSportif": conseiller["IdSportifConseiller"]},{"Nom":1,"IdSportif":1,"_id":0})
        print(request)
```

## 16)	Pour le sportif “Kervadec” quel est le nom de son conseiller ?

```python
Kerva = sportifs.find_one({"Nom":"KERVADEC"})

Conseil = sportifs.find_one({"IdSportif":Kerva["IdSportifConseiller"]})
print(Conseil["Nom"],"est le conseiller de Kervadec")
```

## 17)	Quels entraîneurs entraînent du hand ball et du basket ball ?

```python
for sportif in sportifs.find({"Sports.Entrainer":{"$all":["Hand ball","Basket ball"]}}):
    pprint(sportif,sort_dicts=False)
```
